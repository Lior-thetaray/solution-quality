```python
import os
from datetime import datetime

from thetaray.api.context import init_context
from thetaray.api.distribution import distribute_alerted_activities

spark_conf={
    "spark.executor.memory": "2g",
    "spark.executor.cores": "4",
    "spark.executor.instances": "4",
}

context = init_context(execution_date=datetime(1970, 1, 1), spark_conf=spark_conf)
```

```python
force = context.parameters["mode"] != "test" if "mode" in context.parameters else None
shared_ns = os.environ["SHARED_NAMESPACE"].replace("-", "_")
default_ic_target = shared_ns.replace("shared_", "apps_")

distribute_alerted_activities(context, "tr_analysis", default_ic_target, force=force)
```

```python
context.close()
```
