```python
# Imports
import boto3
import os
from thetaray.common import Settings
from thetaray.common import *
```

```python
from thetaray.api.context import init_context
import datetime
context = init_context(execution_date=datetime.datetime(1970,1,1))
```

```python
boto_session = boto3.Session(aws_access_key_id=Settings.S3_ACCESS_KEY,
                                aws_secret_access_key=Settings.S3_SECRET_KEY)
s3_client = boto_session.client("s3", endpoint_url=Settings.S3_ENDPOINT, verify="/certs/ca.crt")
```

```python
def copy_file(source, connector):
    with open('/thetaray/data/bank/' + source, 'rb') as fp:
        s3_client.upload_fileobj(fp, Settings.S3_PUBLIC_BUCKET, 
                                 os.path.join(Settings.S3_DATASOURCES_PATH, connector, source))


copy_file("trans_En.csv", "transaction")
copy_file("account_En_ER.csv", "account")
copy_file("card.csv", "card")
copy_file("client.csv", "client")
copy_file("disp.csv", "disp")
copy_file("district.csv", "district")
copy_file("loan.csv", "loan")
copy_file("order_En.csv", "order")
copy_file("country_risk.csv", "country_risk")
```
